```python
import pandas as pd

tabela = pd.read_csv('Dep.Federal-2014.csv', sep=';')
tabela.columns = tabela.columns.str.strip()

def transform_foto(foto):
    return foto[20:-11]  # Remove os primeiros 18 e os últimos 11 caracteres
    
# Aplica a função à coluna 'Foto' e renomeia a coluna para 'UF'

tabela['UF'] = tabela['Foto'].apply(transform_foto)

tabela = tabela.drop('Foto', axis=1)

display(tabela.info())
display(tabela)
```

```python
# Dicionário de mapeamento de códigos para nomes completos
uf_mapping = {
    'AC': 'AC',
    'AL': 'AL',
    'AM': 'AM',
    'AP': 'AP',
    'BA': 'BA',
    'CE': 'CE',
    'DF': 'DF',
    'ES': 'ES',
    'GO': 'GO',
    'MA1': 'MA',
    'MG1': 'MG',
    'MS1': 'MS',
    'MT1': 'MT',
    'PA1': 'PA',
    'PB1': 'PB',
    'PE1': 'PE',
    'PI1': 'PI',
    'PR1': 'PR',
    'RJ1': 'RJ',
    'RN2': 'RN',
    'RO2': 'RO',
    'RR2': 'RR',
    'RS2': 'RS',
    'SC2': 'SC',
    'SE2': 'SE',
    'SP2': 'SP',
    'TO2': 'TO',
    'BR2': 'BR'
}

# Aplicar o mapeamento à coluna 'UF'
tabela['UF'] = tabela['UF'].map(uf_mapping)

display(tabela['UF'].unique())
```

```python
ufs = list(tabela['UF'].unique())
print(ufs)

for uf in ufs:
    # Filtrar os dados com base na UF
    dados_uf = tabela[tabela['UF'] == uf]
    
    # Nome do arquivo Excel
    nome_arquivo = f'relatorio{uf}.xlsx'
    
    # Salvar os dados em um arquivo Excel
    dados_uf.to_excel(nome_arquivo, index=False)
```
