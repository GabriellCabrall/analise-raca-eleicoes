```python
import pandas as pd
tabela = pd.read_csv('relatorioAL.csv', sep=';')
tabela.columns = tabela.columns.str.replace('"', '')
tabela["Código Classificador"] = tabela["Código Classificador"].str.replace('"', '')
tabela["Código Classificador"] = tabela["Código Classificador"].astype(int)
tabela = tabela.rename(columns={" Código da Cor": "Código da Cor"})
print(tabela.columns)
tabela["Código da Cor"] = tabela["Código da Cor"].str.replace('"', '')
tabela["Código da Cor"] = tabela["Código da Cor"].astype(int)
#print(tabela['Código Classificador'])
display(tabela) 
# print(tabela.info)
```

```python
# obter os valores únicos da coluna "Código Classificador"
valores_unicos = tabela["Código Classificador"].unique()


# criar uma tabela separada para cada valor único
tabelas_separadas = {}

for valor in valores_unicos:
    tabela_separada = tabela[tabela["Código Classificador"] == valor]
    tabelas_separadas[valor] = tabela_separada

for valor, tabela_separada in tabelas_separadas.items():
    total_linhas = len(tabela_separada)
    serie_contagem = tabela_separada[" Cor"].value_counts()
    serie_incidencia = serie_contagem.div(total_linhas).mul(100)
    print(f"Incidência da cor na tabela para o valor {valor}:")
    print('{}'.format(serie_incidencia))
```
